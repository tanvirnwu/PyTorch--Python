```python
import os
import torch
import torch.nn.functional as F
from torchvision import transforms
from PIL import Image

# Define the paths for the original images folder and the output images folder
original_images_folder = r"C:\Users\tanvi\OneDrive\Desktop\New folder\Papers\Image Dehazing\Datasets\HD DATASET\Sample1"
output_folder = r"C:\Users\tanvi\OneDrive\Desktop\New folder\Papers\Image Dehazing\Datasets\HD DATASET\Sample2"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Define a function to add foggy effect to an image
def add_fog(image_path, fog_intensity):
    # Load the image
    img = Image.open(image_path)

    # Define the foggy effect parameters
    fog_intensity = fog_intensity  # Adjust this value to control the intensity of the fog

    # Apply the foggy effect to the image
    img_tensor = transforms.ToTensor()(img)
    fog_tensor = torch.rand_like(img_tensor) * fog_intensity
    fog_img_tensor = img_tensor + fog_tensor
    fog_img_tensor = torch.clamp(fog_img_tensor, 0, 1)

    # Convert the tensor back to an image
    fog_img = transforms.ToPILImage()(fog_img_tensor)

    return fog_img

# Define a function to add smoky effect to an image
def add_smoke(image_path, smoke_intensity):
    # Load the image
    img = Image.open(image_path)

    # Define the smoky effect parameters
    smoke_intensity = smoke_intensity  # Adjust this value to control the intensity of the smoke

    # Apply the smoky effect to the image
    img_tensor = transforms.ToTensor()(img)
    smoke_tensor = torch.rand_like(img_tensor) * smoke_intensity
    smoke_img_tensor = img_tensor + smoke_tensor
    smoke_img_tensor = torch.clamp(smoke_img_tensor, 0, 1)

    # Convert the tensor back to an image
    smoke_img = transforms.ToPILImage()(smoke_img_tensor)

    return smoke_img

# Define a function to add hazy effect to an image
def add_haze(image_path, hazy_intensity):
    # Load the image
    img = Image.open(image_path, hazy_intensity)

    # Define the hazy effect parameters
    hazy_intensity = hazy_intensity  # Adjust this value to control the intensity of the haze

    # Apply the hazy effect to the image
    img_tensor = transforms.ToTensor()(img)
    haze_tensor = torch.rand_like(img_tensor) * hazy_intensity
    hazy_img_tensor = img_tensor + haze_tensor
    hazy_img_tensor = torch.clamp(hazy_img_tensor, 0, 1)

    # Convert the tensor back to an image
    hazy_img = transforms.ToPILImage()(hazy_img_tensor)

    return hazy_img


# Define a function to add hazy light effect to an image
def add_hazy_light(image_path, hazy_light_intensity):
    # Load the image
    img = Image.open(image_path)

    # Define the hazy light effect parameters
    hazy_light_intensity = 0.5  # Adjust this value to control the intensity of the hazy light

    # Apply the hazy light effect to the image
    img_tensor = transforms.ToTensor()(img)
    hazy_light_tensor = torch.rand_like(img_tensor) * hazy_light_intensity
    hazy_light_img_tensor = img_tensor + hazy_light_tensor
    hazy_light_img_tensor = torch.clamp(hazy_light_img_tensor, 0, 1)

    # Convert the tensor back to an image
    hazy_light_img = transforms.ToPILImage()(hazy_light_img_tensor)

    return hazy_light_img


# Define a function to increase image contrast
def increase_contrast(image_path, contrast_factor):
    # Load the image
    img = Image.open(image_path)

    # Define the contrast enhancement factor
    contrast_factor = contrast_factor  # Adjust this value to control the contrast enhancement

    # Apply the contrast enhancement to the image
    enhancer = transforms.ColorJitter(contrast=contrast_factor)
    contrast_img = enhancer(img)

    return contrast_img

```

```python
import os
import torch
import torch.nn.functional as F
from torchvision import transforms
from PIL import Image, ImageEnhance
import random

# Define the paths for the original images folder and the output images folder
original_images_folder = r"C:\Users\tanvi\OneDrive\Desktop\New folder\Papers\Image Dehazing\Datasets\HD DATASET\Sample1"
output_folder = r"C:\Users\tanvi\OneDrive\Desktop\New folder\Papers\Image Dehazing\Datasets\HD DATASET\Sample2"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Define a function to add a light layer with natural noise to a random region of the image
def add_light_layer(image_path):
    # Load the image
    img = Image.open(image_path)

    # Define the light layer parameters
    light_intensity = 0.2  # Adjust this value to control the intensity of the light layer
    noise_intensity = 0.5  # Adjust this value to control the intensity of the noise

    # Randomly select the region for the light layer
    width, height = img.size
    x1 = random.randint(0, width // 2)
    y1 = random.randint(0, height // 2)
    x2 = x1 + (width // 2)
    y2 = y1 + (height // 2)

    # Create a white image with the same size as the original image
    light_img = Image.new("RGB", img.size, (255, 255, 255))

    # Blend the light layer with the original image in the selected region
    blended_img = Image.blend(img, light_img, light_intensity)

    # Create a copy of the blended image
    noisy_img = blended_img.copy()

    # Add natural noise to the selected region of the blended image
    for y in range(y1, y2):
        for x in range(x1, x2):
            r, g, b = blended_img.getpixel((x, y))
            noise_r = random.uniform(-noise_intensity * 255, noise_intensity * 255)
            noise_g = random.uniform(-noise_intensity * 255, noise_intensity * 255)
            noise_b = random.uniform(-noise_intensity * 255, noise_intensity * 255)
            noisy_img.putpixel((x, y), (int(r + noise_r), int(g + noise_g), int(b + noise_b)))

    # Blend the noisy image with the original image
    final_img = Image.blend(img, noisy_img, noise_intensity)

    return final_img

# Iterate through the original images folder and add a light layer with natural noise to random regions of each image
for filename in os.listdir(original_images_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        input_image_path = os.path.join(original_images_folder, filename)

        # Adding the light layer with natural noise to the image
        light_layer_output_image = add_light_layer(input_image_path)

        # Save the image with the light layer to the output folder
        output_path = os.path.join(output_folder, f"LightLayer_{filename}")
        light_layer_output_image.save(output_path)

print("Images with random light layer added and saved successfully!")
```
